```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('results_p1.csv')
df2 = pd.read_csv('results_p2.csv')

df1 = df1.drop(26)
df2 = df2.drop(26)
df2
```

# P = 1

```python
plt.figure(figsize=(12, 8))

plt.errorbar(df1['dataset'], df1['avg_silhouette1'], yerr=df1['std_silhouette1'], fmt='o', label='Algoritmo 1', capsize=5)
plt.errorbar(df1['dataset'], df1['avg_silhouette2'], yerr=df1['std_silhouette2'], fmt='o', label='Algoritmo 2', capsize=5)
plt.errorbar(df1['dataset'], df1['avg_silhouette_sklearn'], yerr=df1['std_silhouette_sklearn'], fmt='o', label='KMeans (Scikit-learn)', capsize=5)

plt.xlabel('Dataset')
plt.ylabel('Silhouette Score')
plt.title('Comparação do Silhouette Score')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.plot(df1['dataset'], df1['avg_time1'], marker='o', label='Algoritmo 1')
plt.plot(df1['dataset'], df1['avg_time2'], marker='o', label='Algoritmo 2')
plt.plot(df1['dataset'], df1['avg_time_sklearn'], marker='o', label='KMeans (Scikit-learn)')

plt.fill_between(df1['dataset'], df1['avg_time1'] - df1['std_time1'], df1['avg_time1'] + df1['std_time1'], alpha=0.2)
plt.fill_between(df1['dataset'], df1['avg_time2'] - df1['std_time2'], df1['avg_time2'] + df1['std_time2'], alpha=0.2)
plt.fill_between(df1['dataset'], df1['avg_time_sklearn'] - df1['std_time_sklearn'], df1['avg_time_sklearn'] + df1['std_time_sklearn'], alpha=0.2)

plt.xlabel('Dataset')
plt.ylabel('Tempo de Execução (s)')
plt.title('Comparação do Tempo de Execução - p=1')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='n_clusters', y='avg_silhouette1', data=df1, label='Algoritmo 1', marker='o')
sns.lineplot(x='n_clusters', y='avg_silhouette2', data=df1, label='Algoritmo 2', marker='o')
sns.lineplot(x='n_clusters', y='avg_silhouette_sklearn', data=df1, label='KMeans (Scikit-learn)', marker='o')

plt.title('Relação entre Número de Clusters e Silhouette Score')
plt.xlabel('Número de Clusters')
plt.ylabel('Silhouette Score Médio')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='n_clusters', y='avg_adjusted_rand1', data=df1, label='Algoritmo 1', marker='o')
sns.lineplot(x='n_clusters', y='avg_adjusted_rand2', data=df1, label='Algoritmo 2', marker='o')
sns.lineplot(x='n_clusters', y='avg_adjusted_rand_sklearn', data=df1, label='KMeans (Scikit-learn)', marker='o')

plt.title('Relação entre Número de Clusters e Adjusted Rand Index')
plt.xlabel('Número de Clusters')
plt.ylabel('Adjusted Rand Index Médio')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='n_clusters', y='avg_time1', data=df1, label='Algoritmo 1', marker='o')
sns.lineplot(x='n_clusters', y='avg_time2', data=df1, label='Algoritmo 2', marker='o')
sns.lineplot(x='n_clusters', y='avg_time_sklearn', data=df1, label='KMeans (Scikit-learn)', marker='o')

plt.title('Relação entre Número de Clusters e Tempo de Execução')
plt.xlabel('Número de Clusters')
plt.ylabel('Tempo Médio de Execução (s)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

plt.errorbar(df2['dataset'], df2['avg_silhouette1'], yerr=df2['std_silhouette1'], fmt='o', label='Algoritmo 1', capsize=5)
plt.errorbar(df2['dataset'], df2['avg_silhouette2'], yerr=df2['std_silhouette2'], fmt='o', label='Algoritmo 2', capsize=5)
plt.errorbar(df2['dataset'], df2['avg_silhouette_sklearn'], yerr=df2['std_silhouette_sklearn'], fmt='o', label='KMeans (Scikit-learn)', capsize=5)

plt.xlabel('Dataset')
plt.ylabel('Silhouette Score')
plt.title('Comparação do Silhouette Score')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
